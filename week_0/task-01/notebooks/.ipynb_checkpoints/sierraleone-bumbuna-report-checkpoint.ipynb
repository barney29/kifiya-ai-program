```python
import pandas as pd
```

```python
sier_df = pd.read_csv("../data/sierraleone-bumbuna.csv")
sier_df
```

```python
#renaming important columns to avoid confusion 
rename_col = {'GHI': 'global_horizontal_irradiance', 'DNI': 'direct_normal_irradiance', 'DHI': 'diffuse_horizontal_irradiance', 'Tamb': 'ambient_temperature','RH': 'relative_humidity', 'WS': 'wind_speed', 'WD': 'wind_direction', 'TModA':'temp_module_A', 'TModB': 'temp_module_B'}
```

```python
sier_df = sier_df.rename(columns = rename_col)
#inspecting result
sier_df.head()
```

```python
sier_df.shape
```

```python
sier_df.info()
```

```python
sier_df = sier_df.drop_duplicates()
sier_df.count()
```

```python
#droping column Comments 
sier_df = sier_df.drop(['Comments'], axis=1)
```

```python
#cleaning negative values I have used 0 to replace the value
sier_df['global_horizontal_irradiance'] = sier_df['global_horizontal_irradiance'].apply(lambda ghi: 0 if ghi < 0 else ghi)
sier_df['diffuse_horizontal_irradiance'] = sier_df['diffuse_horizontal_irradiance'].apply(lambda dhi: 0 if dhi < 0 else dhi)
sier_df['direct_normal_irradiance'] = sier_df['direct_normal_irradiance'].apply(lambda dni: 0 if dni < 0 else dni)
```

```python
#checking the cleand data
sier_df.sample(10)
```
