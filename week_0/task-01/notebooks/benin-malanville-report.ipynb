## Understanding the given variables

    • Timestamp (yyyy-mm-dd hh:mm): Date and time of each observation.
    • GHI (W/m²): Global Horizontal Irradiance, the total solar radiation received per square meter on a horizontal surface.
    • DNI (W/m²): Direct Normal Irradiance, the amount of solar radiation received per square meter on a surface perpendicular to the rays of the sun.
    • DHI (W/m²): Diffuse Horizontal Irradiance, solar radiation received per square meter on a horizontal surface that does not arrive on a direct path from the sun.
    • ModA (W/m²): Measurements from a module or sensor (A), similar to irradiance.
    • ModB (W/m²): Measurements from a module or sensor (B), similar to irradiance.
    • Tamb (°C): Ambient Temperature in degrees Celsius.
    • RH (%): Relative Humidity as a percentage of moisture in the air.
    • WS (m/s): Wind Speed in meters per second.
    • WSgust (m/s): Maximum Wind Gust Speed in meters per second.
    • WSstdev (m/s): Standard Deviation of Wind Speed, indicating variability.
    • WD (°N (to east)): Wind Direction in degrees from north.
    • WDstdev: Standard Deviation of Wind Direction, showing directional variability.
    • BP (hPa): Barometric Pressure in hectopascals.
    • Cleaning (1 or 0): Signifying whether cleaning (possibly of the modules or sensors) occurred.
    • Precipitation (mm/min): Precipitation rate measured in millimeters per minute.
    • TModA (°C): Temperature of Module A in degrees Celsius.
    • TModB (°C): Temperature of Module B in degrees Celsius.
    • Comments: This column is designed for any additional notes.

## Data Exploration and Data Cleaning

```python
benin_df.head().T
```

```python
#renaming important columns to avoid confusion 
rename_col = {'GHI': 'global_horizontal_irradiance', 'DNI': 'direct_normal_irradiance', 'DHI': 'diffuse_horizontal_irradiance', 'Tamb': 'ambient_temperature','RH': 'relative_humidity', 'WS': 'wind_speed', 'WD': 'wind_direction', 'TModA':'temp_module_A', 'TModB': 'temp_module_B'}
```

```python
benin_df = benin_df.rename(columns = rename_col)
#inspecting result
benin_df.head()
```

```python
benin_df.shape #shape of the dataset
```

```python
benin_df.info()
```

```python
benin_df = benin_df.drop_duplicates()
benin_df.count()
```

```python
benin_df.isnull().sum() #checking for null values
```

```python
#droping column Comments 
benin_df = benin_df.drop(['Comments'], axis=1)
```

```python
#cleaning negative values I have used 0 to replace the value
benin_df['global_horizontal_irradiance'] = benin_df['global_horizontal_irradiance'].apply(lambda ghi: 0 if ghi < 0 else ghi)
benin_df['diffuse_horizontal_irradiance'] = benin_df['diffuse_horizontal_irradiance'].apply(lambda dhi: 0 if dhi < 0 else dhi)
benin_df['direct_normal_irradiance'] = benin_df['direct_normal_irradiance'].apply(lambda dni: 0 if dni < 0 else dni)
```

```python
#checking the cleand data
benin_df.sample(10)
```

```python
benin_df.describe()
```

```python
benin_df['Timestamp'] = pd.to_datetime(benin_df['Timestamp']) # changing to dattime format
```

```python
daily_mean = benin_df.resample('D', on='Timestamp').mean()
daily_mean.reset_index(inplace=True)
daily_mean
```

## Time Series Analysis

### Individual Time Series Analysis

```python
#benin_df.index
```

```python
#To plot a time series analysis for each variables
def plot_time_series(var, label):
    weekly_mean = benin_df.resample('W', on='Timestamp').mean()
    weekly_mean.reset_index(inplace=True)


    plt.figure(figsize=(10, 6))
    plt.plot(np.array(weekly_mean['Timestamp']), np.array(weekly_mean[var]), marker='o', linestyle='-')
    plt.xlabel('Week')
    plt.ylabel(label)
    plt.title(f'Daily Time Series of {label}')
    plt.grid(True)
    plt.show()
```

## daily analysis

```python

plt.figure(figsize=(10, 6))
plt.plot(np.array(daily_mean['Timestamp']), np.array(daily_mean['global_horizontal_irradiance']), marker='o')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Daily Time Series')
plt.grid(True)
plt.show()

```

##  Weekly analysis

```python

weekly_mean = benin_df.resample('W', on='Timestamp').mean()
weekly_mean.reset_index(inplace=True)


plt.figure(figsize=(10, 6))
plt.plot(np.array(weekly_mean['Timestamp']), np.array(weekly_mean['global_horizontal_irradiance']), marker='o', linestyle='-')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Daily Time Series')
plt.grid(True)
plt.show()
```

```python
#Time series analysis of GHI
plot_time_series('global_horizontal_irradiance', 'Global Horizontal Irradiance')
```

```python
#plot Time series analysis of DNI
plot_time_series('direct_normal_irradiance', 'Direct Normal Irradiance')
```

```python
#plot Time series analysis of DHI
plot_time_series('diffuse_horizontal_irradiance','Diffuse Horizontal Irradiance')
```

### Time Series Analysis

```python

#benin_df.set_index('Timestamp', inplace=True)

# Resample the data to weekly and monthly intervals
weekly_df = benin_df.resample('W').mean()  # Weekly mean
monthly_df = benin_df.resample('M').mean()  # Monthly mean

# Create subplots for each variable
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
fig.suptitle('Time Series for GHI, DNI, DHI, and Tamb')

# Plot GHI
weekly_df['global_horizontal_irradiance'].plot(ax=axes[0, 0], title='Weekly GHI')
monthly_df['global_horizontal_irradiance'].plot(ax=axes[1, 0], title='Monthly GHI')

# Plot DNI
weekly_df['direct_normal_irradiance'].plot(ax=axes[0, 1], title='Weekly DNI')
monthly_df['direct_normal_irradiance'].plot(ax=axes[1, 1], title='Monthly DNI')

# Plot DHI
weekly_df['diffuse_horizontal_irradiance'].plot(ax=axes[0, 0], title='Weekly DHI')
monthly_df['diffuse_horizontal_irradiance'].plot(ax=axes[1, 0], title='Monthly DHI')

# Plot Tamb
weekly_df['ambient_temperature'].plot(ax=axes[0, 1], title='Weekly Tamb')
monthly_df['ambient_temperature'].plot(ax=axes[1, 1], title='Monthly Tamb')

# Customize the plots (add labels, legends, etc.)
for ax in axes.flat:
    ax.set_xlabel('Timestamp')
    ax.set_ylabel('Value')
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
corr_matrix = benin_df[['global_horizontal_irradiance', 'direct_normal_irradiance', 'diffuse_horizontal_irradiance', 'ambient_temperature', 'temp_module_A', 'temp_module_B']].corr()
```
