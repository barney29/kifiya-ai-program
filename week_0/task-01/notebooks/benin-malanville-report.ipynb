## Understanding the given variables

    â€¢ Timestamp (yyyy-mm-dd hh:mm): Date and time of each observation.
    â€¢ GHI (W/mÂ²): Global Horizontal Irradiance, the total solar radiation received per square meter on a horizontal surface.
    â€¢ DNI (W/mÂ²): Direct Normal Irradiance, the amount of solar radiation received per square meter on a surface perpendicular to the rays of the sun.
    â€¢ DHI (W/mÂ²): Diffuse Horizontal Irradiance, solar radiation received per square meter on a horizontal surface that does not arrive on a direct path from the sun.
    â€¢ ModA (W/mÂ²): Measurements from a module or sensor (A), similar to irradiance.
    â€¢ ModB (W/mÂ²): Measurements from a module or sensor (B), similar to irradiance.
    â€¢ Tamb (Â°C): Ambient Temperature in degrees Celsius.
    â€¢ RH (%): Relative Humidity as a percentage of moisture in the air.
    â€¢ WS (m/s): Wind Speed in meters per second.
    â€¢ WSgust (m/s): Maximum Wind Gust Speed in meters per second.
    â€¢ WSstdev (m/s): Standard Deviation of Wind Speed, indicating variability.
    â€¢ WD (Â°N (to east)): Wind Direction in degrees from north.
    â€¢ WDstdev: Standard Deviation of Wind Direction, showing directional variability.
    â€¢ BP (hPa): Barometric Pressure in hectopascals.
    â€¢ Cleaning (1 or 0): Signifying whether cleaning (possibly of the modules or sensors) occurred.
    â€¢ Precipitation (mm/min): Precipitation rate measured in millimeters per minute.
    â€¢ TModA (Â°C): Temperature of Module A in degrees Celsius.
    â€¢ TModB (Â°C): Temperature of Module B in degrees Celsius.
    â€¢ Comments: This column is designed for any additional notes.

## Data Exploration and Data Cleaning

```python
benin_df.head().T
```

```python
#renaming important columns to avoid confusion 
rename_col = {'GHI': 'global_horizontal_irradiance', 'DNI': 'direct_normal_irradiance', 'DHI': 'diffuse_horizontal_irradiance', 'Tamb': 'ambient_temperature','RH': 'relative_humidity', 'WS': 'wind_speed', 'WD': 'wind_direction', 'TModA':'temp_module_A', 'TModB': 'temp_module_B'}
```

```python
benin_df = benin_df.rename(columns = rename_col)
#inspecting result
benin_df.head()
```

```python
benin_df.shape #shape of the dataset
```

```python
benin_df.info()
```

```python
benin_df = benin_df.drop_duplicates()
benin_df.count()
```

```python
benin_df.isnull().sum() #checking for null values
```

```python
#droping column Comments 
benin_df = benin_df.drop(['Comments'], axis=1)
```

```python
#cleaning negative values I have used 0 to replace the value
benin_df['global_horizontal_irradiance'] = benin_df['global_horizontal_irradiance'].apply(lambda ghi: 0 if ghi < 0 else ghi)
benin_df['diffuse_horizontal_irradiance'] = benin_df['diffuse_horizontal_irradiance'].apply(lambda dhi: 0 if dhi < 0 else dhi)
benin_df['direct_normal_irradiance'] = benin_df['direct_normal_irradiance'].apply(lambda dni: 0 if dni < 0 else dni)
```

```python
#checking the cleand data
benin_df.sample(10)
```

```python
benin_df.describe()
```

```python
benin_df['Timestamp'] = pd.to_datetime(benin_df['Timestamp']) # changing to dattime format
```

## Time Series Analysis

### Individual Time Series Analysis

```python
#benin_df.index
```

```python
#To plot a time series analysis for each variables
def plot_time_series(var, label):
    plt.figure(figsize=(12, 6))
    benin_df[var].plot()
    plt.xlabel('Timestamp')
    plt.ylabel(label)
    plt.title(f'Time Series Analysis of {label}')
    plt.show()
```

```python
#Time series analysis of GHI
plot_time_series('global_horizontal_irradiance', 'Global Horizontal Irradiance')
```

```python
#plot Time series analysis of DNI
plot_time_series('direct_normal_irradiance', 'Direct Normal Irradiance')
```

```python
#plot Time series analysis of DHI
plot_time_series('diffuse_horizontal_irradiance','Diffuse Horizontal Irradiance')
```

```python
#plotting the variabales over time
plt.figure(figsize=(20, 10))
benin_df[['global_horizontal_irradiance', 'direct_normal_irradiance', 'diffuse_horizontal_irradiance', 'ambient_temperature']].plot()
plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.title('Time Series Analysis of GHI, DNI, DHI and Tamb')
plt.show()
```

```python
corr_matrix = benin_df[['global_horizontal_irradiance', 'direct_normal_irradiance', 'diffuse_horizontal_irradiance', 'ambient_temperature', 'temp_module_A', 'temp_module_B']].corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Solar Radiation Components and Temperature Measures')
```

```python
#wind analysis

sns.pairplot(benin_df[['wind_speed', 'WSgust', 'WSstdev', 'wind_direction', 'WDstdev']])
plt.suptitle('Pairplot Of Wind Speed and Wind Direction Variables')
plt.show()
```

```python
corr_matrix_wind = benin_df[['wind_speed', 'WSgust', 'WSstdev', 'wind_direction', 'WDstdev']].corr()
```

```python
grouped_data = benin_df.groupby('wind_speed')
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix_wind, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Wind Speed and Wind Direction Variables')
plt.show()
```

```python
#correlation
condition = 'wind_speed'
grouped_data = benin_df.groupby(condition)
for condition, group in grouped_data:
    print(f"Correlation matrix for wind speed {condition}")
    correlation_matrix = group[['temp_module_A', 'temp_module_B', 'ambient_temperature']].corr()
    print(correlation_matrix)
    print()
```
